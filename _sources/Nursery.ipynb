## Data Understanding

Data understanding merupakan sebuah tahapan awal di dalam metodologi sains data dan pengembangan AI yang bertujuan untuk mendapatkan pemahaman awal mengenai data yang dibutuhkan untuk memecahkan permasalahan yang ada. Dimana, sering kali data dari berbagai sumber belum tentu dapat langsung digunakan karena:
1. Maksud dan tujuan data berbeda-beda
2. Keadaan asal terpisah-pisah atau justru terintegrasi secara ketat
3. Tingkat kekayaan (richness) berbeda-beda
4. Tngkat keandalan (reliability) berbeda-beda

Langkah awal dalam Data Understanding adalah mengimpor data ke dalam lingkungan Python. Berikut cara mengimpor Nursery Dataset menggunakan Python:

 Sebelum memulai, pastikan telah install package yang akan digunakan ke dalam program Python Anda. Dengan format <pip install> <nama package yang dibutuhkan>. Misalnya: pip install ucimlrepo. Jadinya, kita akan menggunakan package dari ucimlrepo untuk proses nantinya pada data ini.

Jika sudah, lakukan import dataset yang akan dilakukan proses nantinya dengan contoh penerapannya code adalah sebagai berikut.

```
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
nursery = fetch_ucirepo(id=76) 
  
# data (as pandas dataframes) 
X = nursery.data.features 
y = nursery.data.targets 
  
# metadata 
print(nursery.metadata) 
  
# variable information 
print(nursery.variables) 
```

### Convert Dataset to CSV

Langkah selanjutnya merupakan convert dataset yang semula dari online ke offline(file local) berupa ekstensi .csv. Hal ini berguna untuk berbagai hal, di antaranya untuk back up file jika kedepannya ada kendala pada dataset yang ada, untuk mempermudah pelacakan data yang ada di dalamnya, dan banyak lagi. Berikut ini merupakan implementasi code untuk convert dataset ke ekstensi .csv.

```
import pandas as pd

# Mengubah data ke DataFrame
nursery_df = pd.DataFrame(nursery.data.features)

# Menambahkan kolom target
nursery_df['target'] = nursery.data.targets

# Menyimpan data ke file .csv
nursery_df.to_csv('nursery.csv', index=False)

print("Dataset telah disimpan dalam file 'nursery.csv'")
```

Encoding data adalah langkah penting dalam ilmu data yang melibatkan konversi data kategori (kategorikal) menjadi data numerik. ## Modeling Data

Modeling data adalah langkah penting dalam ilmu data yang melibatkan pembangunan model pembelajaran mesin untuk memprediksi atau menjelaskan target variabel. Manfaat Modeling Data, di antaranya mendapatkan prediksi atau penjelasan dari data, mengidentifikasi pola dan hubungan dalam data, membuat keputusan yang lebih informed berdasarkan data, mengotomatisasi tugas dan proses, dan lain sebagainya. ### Klasifikasi Naive Bayes

Klasifikasi Naive Bayes adalah algoritma pembelajaran mesin yang populer untuk tugas klasifikasi, di mana model memprediksi kelas mana yang paling mungkin untuk suatu data baru. Algoritma ini didasarkan pada Teorema Bayes dan asumsi kemerdekaan fitur bersyarat, yang berarti bahwa nilai satu fitur tidak bergantung pada nilai fitur lain.

Cara Kerja Naive Bayes:
1. Hitung probabilitas apriori: Hitung probabilitas setiap kelas dalam dataset.
2. Hitung probabilitas kondisional: Hitung probabilitas setiap fitur untuk setiap kelas.
3. Terapkan Teorema Bayes: Gunakan Teorema Bayes untuk menghitung probabilitas posterior dari setiap kelas untuk data baru.
4. Prediksi kelas: Pilih kelas dengan probabilitas posterior tertinggi sebagai prediksi untuk data baru.

Di bawah ini merupakan implementasi klasifikasi modeling menggunakan Naive Bayes.

```
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Path dataset
X_train_clf = "nursery_train_encoded.csv"
y_train_clf = "nursery_train_target_encoded.csv"
X_test_clf = "nursery_test_encoded.csv"
y_test_clf = "nursery_test_target_encoded.csv"

# Load data as DataFrames
X_train = pd.read_csv(X_train_clf)
y_train = pd.read_csv(y_train_clf)['target']
X_test = pd.read_csv(X_test_clf)
y_test = pd.read_csv(y_test_clf)['target']

# Create and fit the Naive Bayes classifier
gnb = GaussianNB()
gnb.fit(X_train, y_train)

# Make predictions on the test data
y_pred = gnb.predict(X_test)

# Convert DataFrame to list using values
actual_targets = y_test.values.tolist()
predicted_targets = y_pred.tolist()

print("Actual target values:", actual_targets)
print("Predicted target values:", predicted_targets)
```

Setelahnya, akan menghasilkan akurasi guna dipertimbangkan nantinya untuk proses deployment. Dimana memeroleh akurasi tertinggi dari model lainnya. Untuk pertama ini, klasifikasi Naive Bayes mendapatkan akurasi sebagai berikut.

```
# Menghitung accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.3f}%")
```

### Klasifikasi KNN

Klasifikasi KNN (K-Nearest Neighbors) adalah algoritma pembelajaran mesin yang populer untuk tugas klasifikasi, di mana model memprediksi kelas data baru berdasarkan kelas dari data terdekatnya dalam dataset. Algoritma ini bekerja dengan mencari K tetangga terdekat dari data baru dalam ruang fitur dan menggunakan mayoritas kelas dari tetangga tersebut sebagai prediksi kelas data baru.

Cara Kerja KNN:
1. Tentukan nilai K: Pilih nilai K, yang merupakan jumlah tetangga terdekat yang akan digunakan untuk prediksi.
2. Hitung jarak: Hitung jarak antara data baru dan setiap data dalam dataset.
3. Identifikasi tetangga terdekat: Temukan K data dalam dataset yang memiliki jarak terdekat dengan data baru.
4. Prediksi kelas: Hitung mayoritas kelas dari K tetangga terdekat. Kelas yang paling banyak muncul di antara tetangga terdekat akan menjadi 5. prediksi kelas untuk data baru.

Di bawah ini merupakan implementasi klasifikasi modeling menggunakan KNN.

```
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Path dataset
X_train_path = "nursery_train_encoded.csv"
y_train_path = "nursery_train_target_encoded.csv"
X_test_path = "nursery_test_encoded.csv"
y_test_path = "nursery_test_target_encoded.csv"

# Load data as DataFrames
X_train = pd.read_csv(X_train_path)
y_train = pd.read_csv(y_train_path)['target']
X_test = pd.read_csv(X_test_path)
y_test = pd.read_csv(y_test_path)['target']

# Create and fit the KNN classifier with k=105
knn = KNeighborsClassifier(n_neighbors=105)
knn.fit(X_train, y_train)

# Make predictions on the test data
y_pred = knn.predict(X_test)

# Convert DataFrame to list using values
actual_targets = y_test.values.tolist()
predicted_targets = y_pred.tolist()

print("Actual target values:", actual_targets)
print("Predicted target values:", predicted_targets)
```

Setelahnya, akan menghasilkan akurasi guna dipertimbangkan nantinya untuk proses deployment. Dimana memeroleh akurasi tertinggi dari model lainnya. Untuk kedua ini, klasifikasi KNN mendapatkan akurasi sebagai berikut.

```
# Menghitung accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Akurasi: {accuracy:.2f}%")
```

### Ensemble Learning

Ensemble learning adalah sebuah teknik dalam machine learning di mana beberapa model (biasanya disebut sebagai "base learners" atau "weak learners") digabungkan untuk meningkatkan kinerja prediksi dibandingkan dengan penggunaan model tunggal. Ide dasar di balik ensemble learning adalah bahwa dengan menggabungkan berbagai model yang mungkin memiliki kelemahan yang berbeda, kita dapat menghasilkan model yang lebih kuat dan lebih stabil secara keseluruhan. Berikut ini penjelasan bagian-bagian dari tipe Ensemble Learning yang dipelajari pada mata kuliah ditempuh (Penambangan Data) beserta implementasinya.

- Stacking

Stacking melibatkan penggabungan beberapa model berbeda dan menggunakan model yang lebih tinggi (meta-model) untuk menggabungkan hasil prediksi dari model-model tersebut. Stacking memungkinkan untuk memanfaatkan kekuatan berbagai jenis model. Stacking bekerja dengan melatih dua model:
1. Model Dasar: Melatih beberapa model dasar (misalnya, Naive Bayes, KNN) pada dataset.
2. Model Meta: Melatih model meta (misalnya, Naive Bayes) untuk memprediksi kelas data baru berdasarkan prediksi dari model dasar.

Cara Kerja Stacking:
1. Latih Model Dasar: Latih beberapa model dasar pada dataset dan dapatkan prediksi mereka untuk setiap data.
2. Gabungkan Prediksi: Gabungkan prediksi dari model dasar menjadi dataset baru, di mana setiap data diwakili oleh vektor prediksi dari model dasar.
3. Latih Model Meta: Latih model meta pada dataset prediksi yang digabungkan untuk mempelajari hubungan antara prediksi model dasar dan kelas target.
4. Prediksi Baru: Gunakan model meta yang terlatih untuk memprediksi kelas data baru berdasarkan prediksi dari model dasar.

Di bawah ini merupakan implementasi Ensemble Learning menggunakan Stacking.

```
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import StackingClassifier
from sklearn.metrics import accuracy_score


# Path dataset
X_train_el = "nursery_train_encoded.csv"
y_train_el = "nursery_train_target_encoded.csv"
X_test_el = "nursery_test_encoded.csv"
y_test_el = "nursery_test_target_encoded.csv"

# Load data as DataFrames
X_train = pd.read_csv(X_train_el)
y_train = pd.read_csv(y_train_el)['target']
X_test = pd.read_csv(X_test_el)
y_test = pd.read_csv(y_test_el)['target']

# Base Learner 1: KNN (k=105)
clf1 = KNeighborsClassifier(n_neighbors=105)

# Base Learner 2: Naive Bayes
clf2 = GaussianNB()

# Meta-Classifier: Naive Bayes
meta_clf = GaussianNB()

# Create the StackingClassifier object
stacking_clf = StackingClassifier(estimators=[('clf1', clf1), ('clf2', clf2)], final_estimator=meta_clf)

# Fit the StackingClassifier to the training data
stacking_clf.fit(X_train, y_train)

# Make predictions on the test data using the StackingClassifier
y_pred = stacking_clf.predict(X_test)

# Print actual target values and predicted target values
print("Actual target values:", list(y_test))
print("Predicted target values:", list(y_pred))
```

```
import pickle

# Identify index of "id" column (replace 1 if it's a different index)
id_col_index = 1  

# Select features excluding "id"
X_train_selected = X_train.iloc[:, 1:]  # Select all columns except the first

# Train model with selected features
stacking_clf.fit(X_train_selected, y_train)

# Save the modified model
with open('stacking_model-nursery.pkl', 'wb') as f:
    pickle.dump(stacking_clf, f)

print("StackingClassifier model saved to stacking_model-nursery.pkl (excluding 'id' column)")
```

Setelahnya, akan menghasilkan akurasi guna dipertimbangkan nantinya untuk proses deployment. Dimana memeroleh akurasi tertinggi dari model lainnya. Untuk ketiga ini, Ensemble Learning: Stacking mendapatkan akurasi sebagai berikut.

```
# Menghitung accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Akurasi: {accuracy:.2f}%")
```

- Bagging (Bootstrap Aggregating)

Bagging melibatkan pembuatan beberapa subset dari data pelatihan dengan penggantian (bootstrap) dan melatih model yang berbeda pada setiap subset. Kemudian, hasil prediksi dari model-model ini diambil rata-rata (untuk regresi) atau diambil mayoritas (untuk klasifikasi). Oleh karenanya, study case data ini menggunakan klasifikasi, maka model ini nantinya diambil mayoritas.

Di bawah ini merupakan implementasi Ensemble Learning menggunakan Bagging.

Jika, Base Learner menggunakan k-nearest neighbors:

```
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score

# Path dataset
X_train_el = "nursery_train_encoded.csv"
y_train_el = "nursery_train_target_encoded.csv"
X_test_el = "nursery_test_encoded.csv"
y_test_el = "nursery_test_target_encoded.csv"

# Load data as DataFrames
X_train = pd.read_csv(X_train_el)
y_train = pd.read_csv(y_train_el)['target']
X_test = pd.read_csv(X_test_el)
y_test = pd.read_csv(y_test_el)['target']

# Base Learner: KNN
base_clf = KNeighborsClassifier()

n_estimators_list = [55, 75, 105, 125, 145]

for n_estimators in n_estimators_list:
    # Create BaggingClassifier with KNN base learner and set n_estimators
    bagging_clf = BaggingClassifier(base_estimator=base_clf, n_estimators=n_estimators)

    # Fit model on training data
    bagging_clf.fit(X_train, y_train)

    # Predict on testing data
    y_pred = bagging_clf.predict(X_test)

    # Evaluate model (accuracy)
    accuracy = accuracy_score(y_test, y_pred)

    print(f"\n~n_estimators: {n_estimators}~")
    print(f"Akurasi: {accuracy:.3f}%")
```

Setelahnya dilakukan proses di atas, menghasilkan akurasi guna dipertimbangkan nantinya untuk proses deployment. Dimana memeroleh akurasi tertinggi dari model lainnya. Untuk Ensemble Learning: Bagging dengan Base Learner menggunakan KNN mendapatkan akurasi sesuai output code di atas.

Sedangkan, Base Learner menggunakan Naive Bayes:

```
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score

# Path dataset
X_train_el = "nursery_train_encoded.csv"
y_train_el = "nursery_train_target_encoded.csv"
X_test_el = "nursery_test_encoded.csv"
y_test_el = "nursery_test_target_encoded.csv"

# Load data as DataFrames
X_train = pd.read_csv(X_train_el)
y_train = pd.read_csv(y_train_el)['target']
X_test = pd.read_csv(X_test_el)
y_test = pd.read_csv(y_test_el)['target']

# Base Learner: Naive Bayes
base_clf = GaussianNB()

n_estimators_list = [25, 55, 85, 115, 145]  # Daftar n_estimators

for n_estimators in n_estimators_list:
    # Buat model BaggingClassifier dengan base learner dan n_estimators
    bagging_clf = BaggingClassifier(base_estimator=base_clf, n_estimators=n_estimators)

    # Fit model pada data training
    bagging_clf.fit(X_train, y_train)

    # Prediksi pada data testing
    y_pred = bagging_clf.predict(X_test)

    # Evaluasi model
    accuracy = accuracy_score(y_test, y_pred)

    print(f"\n~n_estimators: {n_estimators}~")
    print(f"Akurasi: {accuracy:.3f}%")
```

Setelahnya dilakukan proses di atas, menghasilkan akurasi guna dipertimbangkan nantinya untuk proses deployment. Dimana memeroleh akurasi tertinggi dari model lainnya. Untuk Ensemble Learning: Bagging dengan Base Learner menggunakan Naive Bayes mendapatkan akurasi sesuai output code di atas.

## Evaluasi Data

Evaluasi data adalah langkah penting dalam ilmu data yang melibatkan penilaian performa model pembelajaran mesin pada data yang belum pernah dilihat sebelumnya. Tujuan utama evaluasi data adalah untuk mengukur Akurasi dimana menentukan seberapa baik model memprediksi nilai yang benar untuk data baru dari berbagai cara pada proses Modeling Data. Sehingga dapat menampilkan Matrix umum yang digunakan untuk mengukur akurasi dari hasil yang ada. Berikut ini merupakan visualisasi data untuk evaluasi data dari proses yang telah dilakukan sebelumnya.

```
import matplotlib.pyplot as plt

# Model Accuracies
model_accuracies = {
    "Bagging (KNN)": 0.446,
    "KNN": 0.52,
    "Naive Bayes": 0.810,
    "Stacking": 0.83,
    "Bagging (Naive Bayes)": 0.814,
}

# Mengurutkan label model dan nilai akurasi berdasarkan urutan yang diinginkan (ascending accuracy)
sorted_items = sorted(model_accuracies.items(), key=lambda item: item[1])  # Sort by accuracy value
sorted_model_labels = [item[0] for item in sorted_items]  # Create list of model labels (sorted by accuracy value)
sorted_accuracy_values = [f"{item[1]:.2f}%" for item in sorted_items]  # Create list of accuracy values (formatted as percentage)

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.barh(sorted_model_labels, sorted_accuracy_values, color='skyblue')
plt.xlabel('Accuracy')
plt.title('Perbandingan Model Accuracy')
plt.gca().invert_yaxis()  # Invert y-axis to display highest accuracy at top

# Display the chart
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
